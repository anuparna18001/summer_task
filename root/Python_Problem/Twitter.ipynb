```python
#This function extracts the tweets with 'userhandle' and dumps the extracted data into a json file
def extract_tweets(userhandle):           
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret) 
    auth.set_access_token(access_key, access_secret) 
    api = tweepy.API(auth) 
    # Extracting first 200 tweets ; 200 : max limit 
    tweets = []
    number_of_tweets=200    
    new_tweets = api.user_timeline(screen_name=userhandle,count=number_of_tweets,exclude_replies=True)
    tweets.extend(new_tweets)
    last_id = tweets[-1].id
    #Extracting More Tweets
    while (True):
        new_tweets = api.user_timeline(screen_name=userhandle,count=number_of_tweets,exclude_replies=True,max_id=last_id-1)
        if len(new_tweets)>0:
            last_id = new_tweets[-1].id-1
            tweets.extend(new_tweets)  
        else:
            break
    #dumping tweets into json file
    tmp = []
    tweet = []
    for t in range(0,len(tweets)):
        tmp = (tweets[t]._json)
        tweet.append(tmp)
    with open('tweets.json', 'w') as outfile:
        json_string = json.dump(tweet,outfile)

extract_tweets("midasIIITD")  
```

```python
#This function reads the tweets from the json file and outputs the required data in tabular format as mentioned in the question
def read_tweets(filename):
    tweet_data = {'Text':[], 'Date&Time':[], 'No_of_Favorites':[], 'No_of_Retweets':[], 'No_of_Images':[]}
    extracted_data = pd.DataFrame(tweet_data)
    with open(filename) as json_file:  
        data = json.load(json_file)
        for tweets in data:
            if 'media' in tweets['entities']:
                m = (len(tweets['entities'].get('media', [])))
            else:
                m = 'None'
            extracted_data = extracted_data.append({'Text': tweets['text'], 'Date&Time': tweets['created_at'], 'No_of_Favorites': tweets['favorite_count'], 'No_of_Retweets' : tweets['retweet_count'], 'No_of_Images': m}, ignore_index=True)
    return(extracted_data)

extracted_tweets = read_tweets('tweets.json')
```
