```python
import pickle
import numpy as np 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import csv
```

# Reading the training and testing data

```python
with open('train_image.pkl', 'rb') as f:
    data_image = pickle.load(f)
```

```python
with open('train_label.pkl', 'rb') as f:
    data_label = pickle.load(f)
```

```python
with open('test_image.pkl', 'rb') as f:
    test_data = pickle.load(f)
```

# Spliting the data into train and validation set

```python
X_train, X_test, y_train, y_test = train_test_split(data_image, data_label, test_size=0.2)
```

# Detecting Outliers in the data

```python
#Outlier Detection
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=3, min_samples=2).fit(data_image)
print(clustering.labels_)

#all are clustered in the same cluster, hence no outliers in the data
```

# Applying NN on the data

```python
#Before applying scaling Accuracy 
#Classification with 15 hidden layers
clf0 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(15,))
clf0.fit(X_train, y_train)
y_pred = clf0.predict(X_test)
c=0
ic=0
for i in range(0,len(y_pred)):
    if y_pred[i]==y_test[i]:
        c=c+1
    else:
        ic=ic+1
print('Accuracy : ',c/(c+ic))
```

```python
#Before applying scaling Accuracy 
#Classification with 15 hidden layers
clf6 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(25,))
clf6.fit(X_train, y_train)
y_pred = clf6.predict(X_test)
c=0
ic=0
for i in range(0,len(y_pred)):
    if y_pred[i]==y_test[i]:
        c=c+1
    else:
        ic=ic+1
print('Accuracy : ',c/(c+ic))
```

# Printing data before Scaling

```python
#Data before scaling
print(np.array(X_train).shape)
i=0
for t in X_train:
    new_img = np.array(t).reshape(28,28)
    plt.imshow(new_img,cmap='gray', interpolation = 'bicubic')
    plt.show()
    if i==5:
        break
    i=i+1
```

# Applying Normalization to the data

```python
#scaling
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Data after scaling
print(np.array(X_train).shape)
i=0
for t in X_train:
    new_img = np.array(t).reshape(28,28)
    plt.imshow(new_img, cmap = 'gray', interpolation = 'bicubic')
    plt.show()
    if i==5:
        break
    i=i+1
```
